# Predict whether an employee would be absent on a particular day


```python
# Importing Libraries
import pandas as pd
import numpy as np

raw_data = pd.read_csv('Absenteeism_data.csv')
```

```python
df = raw_data.copy()
```

```python
df.head(10)
```

```python
df = df.drop(['ID'], axis = 1)
df.head()
```

```python
df.describe()
```

```python
df.info()
# No Null Values
```

```python
df['Reason for Absence'].unique()
```

```python
reason_columns = pd.get_dummies(df['Reason for Absence'], drop_first = True)
```

```python
df = df.drop(['Reason for Absence'], axis = 1)
```

```python
reason_1 = reason_columns.loc[:, 1:14].max(axis=1)
reason_2 = reason_columns.loc[:, 15:17].max(axis=1)
reason_3 = reason_columns.loc[:, 18:21].max(axis=1)
reason_4 = reason_columns.loc[:, 22:].max(axis=1)

df = pd.concat([df, reason_1, reason_2, reason_3, reason_4], axis=1)
```

```python
df.head()
```

```python
column_names = ['Date', 'Transportation Expense', 'Distance to Work', 'Age', 'Daily Work Load Average', 'BMI', 'Education', 'Children', 'Pets', 'Absenteeism Time in Hours', 'Reason_1', 'Reason_2', 'Reason_3', 'Reason_4']
df.columns = column_names
```

```python
df.head()
```

```python
column_names_reordered = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Date', 'Transportation Expense', 'Distance to Work', 'Age', 'Daily Work Load Average', 'BMI', 'Education', 'Children', 'Pets', 'Absenteeism Time in Hours', ]

df=df[column_names_reordered]

df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y')
```

```python
list_months = []
```

```python
for i in range(df.shape[0]):
    list_months.append(df['Date'][i].month)

df['Month Value'] = list_months
```

```python
def date_to_weekday(date_value):
    return date_value.weekday()

df['Day of the Week'] = df['Date'].apply(date_to_weekday)
```

```python
df = df.drop(['Date'], axis = 1)

df.head()
```

```python
df['Education'] = df['Education'].map({1:0, 2:1, 3:1, 4:1})
```

```python
df_preprocessed = df.copy()
```

```python
df['Absenteeism Time in Hours'].median()
```

```python
targets = np.where(df['Absenteeism Time in Hours'] > df['Absenteeism Time in Hours'].median(),1,0)
```

```python
targets
```

```python
targets.sum()/ targets.shape[0]
```

```python
targets.shape[0]
```

## Standardisation and train-test split

```python
data_train = df.drop(['Absenteeism Time in Hours'], axis = 1)
```

```python
data_train.head()
```

```python
from sklearn.preprocessing import StandardScaler
absenteeism_scaler = StandardScaler()
```

```python
unscaled_inputs = data_train
```

```python
scaled_inputs = absenteeism_scaler.fit_transform(unscaled_inputs)
```

```python
scaled_inputs
```

```python
scaled_inputs.shape
```

```python
# Train-Test Split

from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(scaled_inputs, targets, train_size = 0.8, random_state = 20)
```

```python
print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

## Logisctic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
reg = LogisticRegression()
```

```python
reg.fit(x_train, y_train)
```

```python
reg.score(x_train, y_train)
```

# Manually Checking the accuracy
